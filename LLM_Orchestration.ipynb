```python
question = "Please come up with a extremely challenging, nuanced question that I can ask a number of LLMs to evaluate their intelligence level."

question += "Answer with only one question! No explanations!"

messages = [{"role" : "user", "content" : question}]
```

```python
from groq import Groq

groq_client = Groq()

host_llm_response = groq_client.chat.completions.create(

                                                   model = 'llama-3.3-70b-versatile',

                                                   messages = messages,

)

host_llm_response_question = host_llm_response.choices[0].message.content

print(display(Markdown(host_llm_response_question)))
```

```python
competitors = []

answers = []
```

```python
messages_for_groq_llama70b = [{"role" : "user", "content" : host_llm_response_question }]

model_name1 = 'llama-3.3-70b-versatile'

llama70b_response = groq_client.chat.completions.create(

                                                                        model = model_name1,

                                                                        messages = messages_for_groq_llama70b
)

llama70b_response = llama70b_response.choices[0].message.content

print(display(Markdown(llama70b_response)))

competitors.append(model_name1)

answers.append(llama70b_response)
```

```python
from google import genai

gemini_client = genai.Client()

messages = host_llm_response_question

model_name2 = 'gemini-2.0-flash'

contestant_llm_1_response = gemini_client.models.generate_content(

                                                             model = model_name2,

                                                             contents = messages

)

gemini_response = contestant_llm_1_response.text

print(display(Markdown(gemini_response)))

competitors.append(model_name2)

answers.append(gemini_response)
```

```python
print(competitors)

print(answers)
```

```python
for competitor, answer in zip(competitors, answers):

  print(f"Competitor: {competitor}\n\n{answer}")
```

```python
answers_combined = ""

for idx, answer in enumerate(answers):

  answers_combined += f"# Response from competitor{idx + 1} \n\n"

  answers_combined += answer + "\n\n"

print(answers_combined)
```

```python
judge_llm_task = f"""You are the judge for a competition between {len(competitors)} llm contestants.

Each llm has been asked this question :  {host_llm_response_question}

Now you as a special guest and judge for this competition should evaluate the depthness, clarity and quality of each answer and rank them, from the best one coming on top and so on.

Frame your response strictly in JSON format in the following way:

{{"results": ["1", "2",....]}}

Here are the answers derived and articulated by each competitor:

{answers_combined}

Alright! now respond with the JSON including with the ranked order of the competitors, nothing else. STRICTLY DO NOT INCLUDE MARKDOWN FORMATTING OR CODE BLOCKS!

"""
```

```python
print(judge_llm_task)
```

```python
!pip install openai

from openai import OpenAI

openai_api_key = os.environ['OPENAI_API_KEY'] = userdata.get('OpenAI_API_Key')
```

```python
openai_client = OpenAI()


judge_llm_messages = [{"role" : "user","content" : judge_llm_task}]

judge_llm_result = openai_client.chat.completions.create(

                                                       model = 'gpt-4o-2024-08-06',

                                                       messages = judge_llm_messages
)

competition_result = judge_llm_result.choices[0].message.content

print(display(Markdown(competition_result)))
```

```python
results_dictionary = json.loads(competition_result)

rank = results_dictionary["results"]

for index, result in enumerate(rank):

  competitor = competitors[int(result) - 1]

  print(f"Rank : {index + 1} : {competitor}")
```
